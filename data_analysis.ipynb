```python
# Group by product name
product_group = df.groupby('Product Name')

# Calculate total sales and profit
product_sum = product_group[['Sales', 'Profit']].sum().reset_index()

# Top 10 Products sorted by profit
top_product_sort = product_sum.sort_values(by='Profit', ascending=False).head(10)

print(top_product_sort)

# Graph Plot
plt.figure(figsize=(12, 6))
plt.barh(top_product_sort['Product Name'], top_product_sort['Sales'])
plt.ylabel('Product Name')
plt.xlabel('Total Sales')
plt.title('Top 10 Profitable Products by Sales')


plt.show()
```
